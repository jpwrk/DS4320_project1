```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\jiaro\Downloads\DS4320_project1\creditcard.csv")


# Proof-of-concept: count fraud vs non-fraud transactions
fraud_count = df['Class'].value_counts()
print("\nNumber of transactions by class (0 = non-fraud, 1 = fraud):")
print(fraud_count)

# calculate basic statistics for the transaction amount
amount_stats = df['Amount'].describe()
print("\nSummary statistics for transaction Amount:")
print(amount_stats)

# simple filtering demonstration
large_transactions = df[df['Amount'] > 2000]
print(f"\nNumber of transactions over $2000: {len(large_transactions)}")
```
